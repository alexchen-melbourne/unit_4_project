## Data Cleaning

### Drop duplicate and na

```python
# Load data scraping from au.indeed.com (details in Scraping notebook)
jobs = pd.read_csv('jobs.csv')
jobs.drop('Unnamed: 0', axis=1, inplace=True)
print jobs.shape
jobs.drop_duplicates(inplace=True)
print jobs.shape
jobs.dropna(inplace=True)
print jobs.shape
jobs.columns = ['city', 'title', 'company', 'location', 'description', 'salarytype', 'salary']
```

## Get dummies

```python
jobs['state'] = jobs.location.map(lambda x: x.split()[-1])
jobs.drop('location', axis=1, inplace=True)

state_dummies = pd.get_dummies(jobs.state)
jobs = pd.concat([jobs, state_dummies], axis=1)
jobs.drop(['state', 'Australia'], axis=1, inplace=True)

# city_dummies = pd.get_dummies(jobs.city)
# jobs = pd.concat([jobs, city_dummies], axis=1)
# jobs.drop(['city', 'Newcastle'], axis=1, inplace=True)

type_dummies = pd.get_dummies(jobs.salarytype)
jobs = pd.concat([jobs, type_dummies], axis=1)
jobs.drop(['salarytype', 'month'], axis=1, inplace=True)
```

```python
def f(x):
    x = re.sub(r'\([^)]*\)', '', x)
    return x

jobs.title = jobs.title.map(f)
```

```python
def ifSenior(x):
    if 'senior' in x.lower():
        return 1
    else:
        return 0

jobs['ifSenior'] = jobs.title.map(ifSenior)
```

```python
def ifJunior(x):
    if 'junior' in x.lower():
        return 1
    else:
        return 0

jobs['ifJunior'] = jobs.title.map(ifJunior)
```

```python
snow = SnowballStemmer('english')
port = PorterStemmer()
lanc = LancasterStemmer()

def stem(x):
    line = re.sub(r'[-|,./$&\'+]', ' ', x)
    word = line.split()
    new = ' '.join([snow.stem(i) for i in word])

    return new
jobs['stemTitle'] = jobs.title.map(stem)
```

```python
def stemd(x):
    line = re.sub(r'[-|,./$&\'+\[\]]', ' ', x)
    word = line.split()
    new = ' '.join([lanc.stem(i) for i in word])
    return new

jobs['stemDesc'] = jobs.description.map(stemd)
```

```python
jobs.reset_index(inplace=True)
jobs.drop('index', axis=1, inplace=True)
```

```python
def levelSalary(x):
    if x < 90000:
        return 0
    else:
        return 1
jobs['salarylevel'] = jobs.salary.map(levelSalary)
```

## What distinguish ds from other jobs?

```python
def ifds(x):
    if 'data sci' in x.lower():
        return 1
    else:
        return 0
jobs['ifds'] = jobs.title.map(ifds)

def rmdigit(x):
    result = ''.join([i for i in x if not i.isdigit()])
    return result
descrip = jobs.stemDesc.map(rmdigit)

stop_words = text.ENGLISH_STOP_WORDS.union(['perm','bi'])

cvec = CountVectorizer(stop_words=stop_words)
cvec.fit(descrip)

words = pd.DataFrame(cvec.transform(jobs.title).todense(), columns=cvec.get_feature_names())
words = words.rename(columns = {'fit': 'fit_feature'})
```

```python
X = words
y = jobs.ifds.values

Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.2, random_state=4)
print 'baseline', 1- float(y.sum())/len(y)

from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(max_depth=2, random_state=0)

clf.fit(Xtrain, ytrain)
yhat = clf.predict(Xtest)
print 'RandomForest: ', accuracy_score(ytest, yhat)

logr = LogisticRegression()
logr.fit(Xtrain, ytrain)
yhat2 = logr.predict(Xtest)
print 'Logistic Reg: ', accuracy_score(ytest, yhat2)

skb_f = SelectKBest(f_classif, k=5)
skb_chi2 = SelectKBest(chi2, k=5)

skb_f.fit(X, y)
skb_chi2.fit(X, y)

kbest = pd.DataFrame([list(X.columns), list(skb_f.scores_), list(skb_chi2.scores_)], 
                     index=['feature','f_classif','chi2 score']).T.sort_values('f_classif', ascending=False)
kbest.head(10)
```

# ROC

Your boss would rather tell a client incorrectly that they would get a lower salary job than tell a client incorrectly that they would get a high salary job. Adjust one of your models to ease his mind, and explain what it is doing and any tradeoffs. Plot the ROC curve.

```python
y = jobs.salarylevel.values
X = jobs.iloc[:, 5:17]

X_train, X_test, y_train, y_test = train_test_split(jobs.title, jobs.salarylevel, train_size = .5, random_state=90)

model = make_pipeline(CountVectorizer(stop_words='english'),
                      LogisticRegression())
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
yhat_pp = model.predict_proba(X_test)
```

```python
conmat = np.array(confusion_matrix(y_test, y_pred, labels=[1,0]))

confusion = pd.DataFrame(conmat, index=['is_high', 'is_low'],
                         columns=['predicted_high','predicted_low'])
confusion
```

```python
pp = pd.DataFrame(yhat_pp, columns=['lowSalary','highSalary'])
pp.head()
```

```python
for thresh in np.arange(1,100)/100.:
    labeled_highS = np.array([1 if x >= thresh else 0 for x in pp.highSalary.values])
    print 'Threshold:', thresh, 'false positives:', np.sum((y_test == 0) & (labeled_highS == 1))
```

```python
labeled_highS = np.array([1 if x >= 0.9 else 0 for x in pp.highSalary.values])
print 'Threshold: 0.9 false positives:', np.sum((y_test == 0) & (labeled_highS == 1))
```

```python
conmat = np.array(confusion_matrix(y_test, labeled_highS, labels=[1,0]))

confusion = pd.DataFrame(conmat, index=['is_high', 'is_low'],
                         columns=['predicted_high','predicted_low'])
confusion
```

```python
from sklearn.metrics import roc_curve, auc

fpr, tpr, _ = roc_curve(y_test, yhat_pp[:,1])
roc_auc = auc(fpr, tpr)

plt.figure(figsize=[8,8])
plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc, linewidth=4)
plt.plot([0, 1], [0, 1], 'k--', linewidth=4)
plt.xlim([-0.05, 1.0])
plt.ylim([-0.05, 1.05])
plt.xlabel('False Positive Rate', fontsize=18)
plt.ylabel('True Positive Rate', fontsize=18)
plt.title('Receiver operating characteristic: high salary', fontsize=18)
plt.legend(loc="lower right")
plt.show()
```
